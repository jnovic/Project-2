```python
import pandas as pd
import numpy as np
import csv
import os
import sqlite3
import sqlalchemy
from sqlalchemy import create_engine
```

```python
review_file = "google-play-store-apps/googleplaystore_user_reviews.csv"
review_df = pd.read_csv(review_file)
review_df.head()
```

```python
app_file = "google-play-store-apps/googleplaystore.csv"
app_df = pd.read_csv(app_file)
app_df.head()
```

```python
merge_df = pd.merge(app_df, review_df, on="App")
merge_df.head()
```

```python
drop_columns = ["Size",'Content Rating','Genres','Last Updated', 'Sentiment_Polarity','Sentiment_Subjectivity']
clean_df = merge_df.drop(drop_columns, axis=1)
clean_df.head()
```

```python
clean_df = clean_df.dropna()
clean_df.count()
```

```python
conn = sqlite3.connect("googleapps.sqlite")
```

```python
final_db = clean_df.to_sql("apps", conn)
```

```python
engine = create_engine("sqlite:///googleapps.sqlite")
```

```python
data = engine.execute("SELECT * FROM apps LIMIT 5")
```

```python
for record in data:
    print(record)
```
